# 1D J1J2=0.8: Poincare GRU

This is part of the work https://arxiv.org/abs/2604.24337.

```python
import os
import sys
import time
sys.path.append('../utility')
from j1j2_tau_hyprnn_train_loop import *
```

```python
E_exact = -42.07006297371643
units = 70
syssize = 100
nssamples = 80
J1 = 1.0
J2 = 0.8
nsteps = 1001
var_tol = 20.0
r_max=1.0
fname = f'1d_J1J2_results_N=100_tau_hypGRU/rmax={r_max}/J2={J2}'
```

```python
def set_cpu_deterministic(seed=111):
    # 1. Python & Numpy
    random.seed(seed)
    np.random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    
    # 2. PyTorch CPU
    torch.manual_seed(seed)
    
    # 3. Force Deterministic Algorithms
    # This prevents non-deterministic CPU operations (like some views/reductions)
    torch.use_deterministic_algorithms(True)
    
    # 4. Limit CPU Threads
    # Setting this to 1 ensures operations are done in a fixed order.
    torch.set_num_threads(1)

set_cpu_deterministic(111)
```

# PoincareGRU

```python
hGRU = RNNwavefunction_hyp(systemsize=syssize, cell_type='HypGRU', bias_geom='hyp',
                           hyp_non_lin='id', units=units, r_max=r_max, seed=111)
for name, param in hGRU.model.named_parameters():
    print(f"Layer: {name} | Size: {param.size()}")
    
t0=time.time()
mhE, vhE = run_J1J2(hGRU, nsteps, syssize, var_tol, J1_=J1, J2_=J2, Marshall_sign=True, 
                   numsamples=nssamples, lr1=5e-3, lr2=1e-3, seed=111, fname=fname)
t1=time.time()
print(f'Time taken = {np.round((t1-t0)/3600,3)} hrs')
```
